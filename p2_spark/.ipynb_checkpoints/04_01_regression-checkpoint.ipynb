```python
import pyspark
```

```python
from pyspark.sql import SparkSession, SQLContext
```

```python
spark = SparkSession.builder.getOrCreate()
```

## Linear R

```python
df = spark.read.parquet("processed_hmp_data.parquet/")
```

```python
df.show(5)
```

```python
df.createOrReplaceTempView("df")
```

```python
spark.sql("""
    select class, sqrt(sum(x*x) + sum(y*y) + sum(z*z)) as label
    from df
    group by class
""").show()
```

```python
class_energy = spark.sql("""
    select class, sqrt(sum(x*x) + sum(y*y) + sum(z*z)) as label
    from df
    group by class
""")
```

```python
class_energy.createOrReplaceTempView("energy")
```

```python
df_join = spark.sql("""
    select x, y, z, df.class, energy.label
    from df
    join energy
    on df.class = energy.class
""")
```

```python
df_join.show(5)
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
lr = LinearRegression(maxIter=10, regParam=0.3, elasticNetParam=0.8)
```

```python
from pyspark.ml.feature import VectorAssembler

vectorAssembler = VectorAssembler(inputCols=["x", "y", "z"], outputCol="features")
```

```python
from pyspark.ml import Pipeline
```

```python
pipeline = Pipeline(stages=[vectorAssembler, lr])
```

```python
model = pipeline.fit(df_join)
```

```python
preds = model.transform(df_join)
```

```python
preds.show()
```

```python
pd_df_preds = preds.sample(fraction=0.1).toPandas()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(pd_df_preds[["x", "y", "z", "label", "class", "prediction"]])
```

```python
model.stages[1].summary.r2
```

```python
model.stages[1].summary.r2adj
```

## Logistic R

```python
from pyspark.ml.classification import LogisticRegression
```

```python
logr = LogisticRegression(labelCol="classIndex")
```

```python
df.show()
```

```python
pipeline = Pipeline(stages=[logr])
```

```python
model = pipeline.fit(df)
```

```python
model.stages[0].summary.accuracy
```

```python
prediction = model.transform(df)
```

```python
prediction.show()
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
eval = MulticlassClassificationEvaluator(labelCol='classIndex').setMetricName("accuracy").setLabelCol('classIndex').setPredictionCol('prediction')
```

```python
eval.evaluate(prediction)
```
